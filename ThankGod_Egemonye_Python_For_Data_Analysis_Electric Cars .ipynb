```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# reading CSV file of an EV dataset
Car_Data = pd.read_csv('Electric Cars Vs Motor Cars.csv', sep = ';') 
```

```python
# getting firts 50 data from the dataset
Car_Data.head(50)
```

```python
# statistical data of the dataset
Car_Data.describe()
```

```python
# filling NAN's with 0
Cars=Car_Data.fillna(0)
Cars
```

```python
# Univariate Boxplot for BEV Sales
fig, axs = plt.subplots(figsize = (10,5))
plot =sns.boxplot(x='Year', y ='EV sales BEV', data=Cars, ax=axs)
plot = plot.set_xticklabels(plot.get_xticklabels(), rotation =40);
plt.title('Boxplot Showing BEV Sales Vs Year', size=15)
```

```python
# Univariate Boxplot for PHEV 
fig, axs = plt.subplots(figsize = (10,5))
plot = sns.boxplot(x='Year', y ='EV sales PHEV', data=Cars, ax=axs)
plot.set_xticklabels(plot.get_xticklabels(), rotation =40);
plt.title('Boxplot Showing PHEV Sales Vs Year', size=15)
```

```python
# Histogram plot for BEV Sales
sns.histplot(x =Cars['EV sales BEV'])
```

```python
# Histogram plot for PHEV Sales
sns.histplot(x =Cars['EV sales PHEV'])
```

```python
# Reading another Electric Vehicle Related Dataset
EVs = pd.read_csv('Electric_Vehicle_Population_Data.csv')
EVs.head(50)
```

```python
# Performing data wrangling and cleaning to ensure quality of data and to remove unnecessary data
New_EVs =EVs.drop(columns=['Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'VIN (1-10)', 'Vehicle Location', 'DOL Vehicle ID', 'Postal Code','Base MSRP', 'Legislative District', '2020 Census Tract'])
New_EVs.head(50)
```

```python
# fetching total population of cars by individual car maker 
EVs_1 = New_EVs['Make'].value_counts()
EVs_1
```

```python
# Fetching only data relating to TESLA cars
New_EVs.loc[EVs['Make'] == 'TESLA']
```

```python
# Counting total Electric Utility supply Per company
EVs_2 = New_EVs['Electric Utility'].value_counts()
EVs_2.head(50)
```

```python
# Fetching all the model produced by tesla and total population present in the dataset
EV_df= New_EVs.loc[EVs['Make'] == 'TESLA', ['Model']].value_counts()
EV_df
```

```python
# Fetching Battery Electric Vehicles with range > 300 miles
EV_df2 = New_EVs.loc[(New_EVs['Electric Range'] > 300) & New_EVs[('Electric Vehicle Type')].isin(['Battery Electric Vehicle (BEV)'])]
EV_df2
```

```python
# Fetching Plug-in Hybrid Electric Vehicles with range > 100 miles
EV_df3 = New_EVs.loc[(New_EVs['Electric Range'] > 100) & New_EVs[('Electric Vehicle Type')].isin(['Plug-in Hybrid Electric Vehicle (PHEV)'])]
EV_df3
```

```python
# Fetching maximum and minimum range associated with car models for respective car makers: ist 60 index
Df_EV=New_EVs.groupby(['Make', 'Model'])['Electric Range'].agg(['max','min'])
Df_EV.head(60)
```

```python
# Fetching maximum and minimum range associated with car models for respective car makers: last 60 index
Df_EV=New_EVs.groupby(['Make', 'Model'])['Electric Range'].agg(['max','min'])
Df_EV.tail(60)
```
